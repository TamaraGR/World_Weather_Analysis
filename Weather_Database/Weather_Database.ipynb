```python
# Import dependencies

import pandas as pd
import numpy as np
from citipy import citipy
import requests
from config import weather_api_key
from datetime import datetime
```

```python
# Create a new set of 2,000 random latitudes and longitudes

lats = np.random.uniform(-90.000,90.000, size=2000)
lngs = np.random.uniform(-180.000,180.000, size=2000)

lats_lngs = zip(lats,lngs) 
lats_lngs
```

```python
#Put the lats and lngs to a list

coordinates = list(lats_lngs)
```

```python
#Get the nearest city using the citipy module.

#list for holding the cities

cities = []

#Nearest city for each Lat and Lngs combo.

for coord in coordinates:
    city = citipy.nearest_city(coord[0],coord[1]).city_name
    
    #Every unique city we will add to the cities list.
    
    if city not in cities:
        cities.append(city)    
```

```python
#Get city count to ensure we have enough cities. 

len(cities)
```

```python
#Start URL for the Weather Map API Call. 

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python

# Create an empty list to hold the weather data
city_data = []

# Print the beginning of the logging
print("Beginning of data retrieval          ")
print("-------------------------------------")

# Create the counters and initialize to 1 because the first iteration of logging for each recorded response and
# the set needs to start at 1
record_count = 1
set_count = 1

# Loop through the cities in the list
for i, city in enumerate(cities):
    
    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        
    # Create endpoint URL with each city
    city_url = url + "&q=" + city.replace(" ", "+")
    
    # Log the URL, record, and set numbers and the city
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1 to the record count
    record_count += 1


# Run an API request for each of the cities
    try:
        # Parse the JSON and retrieve the data
        city_weather = requests.get(city_url).json()
        
        # Parse out the needed data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        
        # Convert the date to ISO
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        # Append the city information to the city_data list
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})
        
# If an error is given, skip the city
    except:
        print("City not found.  Skipping...")
        pass
    
# Indicate the date loading is complete
print("------------------------------")
print("Date retrieval complete       ")
print("------------------------------")
```

```python
#Confirm sufficient city count
len(city_data)
```

```python
# Convert dictionaries arrays to a DF
city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```

```python
# CSV output file
output_data_file = "WeatherPy_Database.csv"

# Export city_data into a CSV
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
